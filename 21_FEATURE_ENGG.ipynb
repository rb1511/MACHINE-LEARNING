```python
from sklearn.datasets import make_circles
```

```python
from sklearn.decomposition import KernelPCA
```

```python
X,y = make_circles(n_samples=400,factor=0.3,noise=0.05)
```

```python
X
```

```python
y
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
```

```python
test_df = pd.DataFrame(X,columns=['a','b'])
```

```python
test_df
```

```python
sns.scatterplot(x='a',y='b',hue=y,data=test_df)
```

```python
kpca = KernelPCA(kernel='rbf',gamma=10)
```

```python
X_pca = kpca.fit_transform(X)
```

```python
X_pca
```

```python
X_pca = pd.DataFrame(X_pca)
```

```python
X_pca
```

```python
sns.scatterplot(x=0,y=1,hue=y,data=X_pca)
```

# Feature Selection

```python
import numpy as np
```

```python
from sklearn.datasets import load_iris
```

```python
raw_data = load_iris()
```

```python
raw_data
```

```python
X = raw_data.data
```

```python
X
```

```python
y = raw_data.target
```

```python
y
```

```python
from sklearn.feature_selection import VarianceThreshold
```

```python
sel = VarianceThreshold(threshold=0.6)
```

```python
X_vt = sel.fit_transform(X)
```

```python
X_vt
```

```python
test_df = pd.DataFrame(X_vt,columns=['a','b'])
```

```python
test_df
```

```python
sns.scatterplot(x='a',y='b',hue=y,data=test_df)
```

```python
from sklearn.feature_selection import SelectKBest,SelectPercentile,chi2
```

```python
X_sb = SelectKBest(chi2,k=3).fit_transform(X,y)
```

```python
X_sb
```

```python
X_sp = SelectPercentile(chi2,percentile=75).fit_transform(X,y)
```

```python
X_sp
```
