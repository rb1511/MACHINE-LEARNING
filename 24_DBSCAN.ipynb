```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style('whitegrid')
from sklearn.datasets import make_blobs
```

```python
centers = [[0.5,2],[-1,-1],[1.5,-1]]
```

```python
centers
```

```python
X,y = make_blobs(n_samples=400,centers=centers,cluster_std=0.5,random_state=0)
```

```python
X
```

```python
y
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X = scaler.fit_transform(X)
```

```python
X
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c=y,cmap='rainbow')
plt.show()
```

```python
from sklearn.cluster import DBSCAN
```

```python
db = DBSCAN(eps=0.6,min_samples=100) #eps i.e epsilon
# The DBSCAN algorithm basically requires 2 parameters: eps: 
#specifies how close points should be to each other to be considered a part of a cluster. 
#It means that if the distance between two points is lower or equal to this value (eps), 
#these points are considered neighbors.
```

```python
db.fit(X)
```

```python
identified_cluster = db.fit_predict(X)
```

```python
identified_cluster
```

```python
plt.figure(figsize=(10,6))
plt.scatter(X[:,0],X[:,1],c= identified_cluster,cmap='rainbow')
plt.show()
```
