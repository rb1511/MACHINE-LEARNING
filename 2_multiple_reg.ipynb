```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style ('whitegrid')
```

```python
data = pd.read_csv('multisat.csv')
```

```python
# GPA = b0 + b1*SAT+b2*Rand 123 # y = b0+b1*x1+b2*x2
```

```python
z = data[['SAT','Rand 1,2,3']]
```

```python
z
```

```python
y = data['GPA']
```

```python
y
```

```python
import statsmodels.api as sm
```

```python
x = sm.add_constant(z)
```

```python
x
```

```python
#creat model
results = sm.OLS(y,x).fit()
```
